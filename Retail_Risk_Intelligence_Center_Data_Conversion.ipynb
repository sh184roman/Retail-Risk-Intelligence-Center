```python
import pandas as pd
import glob
import os

# Folder containing parquet files
folder = r"C:\Users\user\Downloads\contoso"

# Get all parquet files
parquet_files = glob.glob(os.path.join(folder, "*.parquet"))

# Convert each parquet file to csv
for file in parquet_files:
    df = pd.read_parquet(file)

    filename = os.path.basename(file)

    # Extract name after "_v" and change extension to .csv
    csv_name = filename.split("_v")[1].replace(".parquet", ".csv")

    csv_file = os.path.join(folder, csv_name)

    df.to_csv(csv_file, index=False)

    print(f"Converted: {filename} -> {csv_name}")

print("All files converted successfully!")
```
